```python
import random
from datetime import datetime, timedelta
import json
import os

dir_files_db = "."

filepath_items = os.path.join(dir_files_db, "db", "mock_data", "items.json")
filepath_items_stocks = os.path.join(
    dir_files_db, "db", "mock_data", "items_stocks.json"
)

items_stocks = {}


# Helper function to generate random dates for history
def random_date(start_date, end_date):
    time_delta = end_date - start_date
    random_days = random.randint(0, time_delta.days)
    return start_date + timedelta(days=random_days)


today = datetime.now()
six_months_ago = today - timedelta(days=180)
one_year_ago = today - timedelta(days=365)

# Assuming items from the previous response is available
with open(filepath_items, "r") as file:
    items: dict = json.load(file)

for item_id, item_data in items.items():
    items_stocks[item_id] = {}

    # Shipment History
    shipment_history = []
    num_shipments = random.randint(0, 3)  # Up to 3 past shipments
    for _ in range(num_shipments):
        ship_date = random_date(one_year_ago, today)
        ship_qty = random.randint(5, 50)
        ship_status = random.choice(["Delivered", "In Transit", "Delayed"])
        ship_destination = random.choice(
            ["Warehouse A", "Warehouse B", "Store A", "Store B"]
        )
        shipment_history.append(
            {
                "date": ship_date.strftime("%Y-%m-%d"),
                "quantity": ship_qty,
                "status": ship_status,
                "destination": ship_destination,
            }
        )
    items_stocks[item_id]["shipments"] = shipment_history

    # Magazine Locations (Assuming "magazine" means "warehouse/storage location")
    locations = []
    num_locations = random.randint(1, 3)
    for _ in range(num_locations):
        warehouse = random.choice(
            ["Main Warehouse", "Regional Depot East", "Regional Depot West"]
        )
        aisle = f"{random.randint(1, 20)}"
        shelf = f"{random.randint(1, 10)}"
        location = dict(
            quantity=random.randint(0, 50),
            location=warehouse,
            aisle=aisle,
            shelf=shelf,
            details=None,
        )
        locations.append(location)
    items_stocks[item_id]["locations"] = locations

with open(filepath_items, "w") as file:
    json.dump(items, file, indent=4)
with open(filepath_items_stocks, "w") as file:
    json.dump(items_stocks, file, indent=4)
```

```python
filepath_customers = os.path.join(dir_files_db, "db", "mock_data", "customers.json")

with open(filepath_customers, "r", encoding="utf-8") as file:
    customers = json.load(file)
    customers = dict(
        (customer.get("cid"), customer)
        for cid, customer in enumerate(customers.values())
    )
with open(filepath_customers, "w") as file:
    json.dump(customers, file, indent=4)


with open(filepath_items, "r", encoding="utf-8") as file:
    items = json.load(file)
    items = dict((iid, {**item, "iid": iid}) for iid, item in items.items())
with open(filepath_items, "w") as file:
    json.dump(items, file, indent=4)

with open(filepath_items_stocks, "r", encoding="utf-8") as file:
    items_stocks = json.load(file)
    items_stocks = dict(
        (iid, {**item, "iid": iid}) for iid, item in items_stocks.items()
    )
with open(filepath_items_stocks, "w") as file:
    json.dump(items_stocks, file, indent=4)
```
