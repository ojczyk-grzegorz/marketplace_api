```python
import duckdb
```

```python
db = duckdb.connect("logs.duckdb")
db
```

```python
db.execute("""
SELECT timestamp, level, log FROM read_json_auto('logs/app.log');
""").fetchall()
```

```python
db.execute("""
SELECT
    timestamp,
    level,
    UNNEST(log)
FROM read_json_auto('logs/app.log');
""").fetchall()
```
